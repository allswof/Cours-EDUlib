## Pratique 1

```python
# Importer pandas
import pandas as pd
```

Ici on vous fournit les lien url pour les fichiers **films.xlsx** et **rating.csv**. En vous aidant des anciens notebooks vu lors de cette formation, complétez les lignes de code afin de procéder au chargement des données. 
Mettre le tout dans deux dataframes: **movies** et **rating**

```python
# charger les données contenues dans un fichiers excel ou csv se trouvant sur drive
URL_films   = 'https://docs.google.com/spreadsheets/d/15mxFMzmEZpS0a-qa0b-ZD-kDmdY2egTH/edit?usp=sharing&ouid=109066890121411090393&rtpof=true&sd=true'
movies_file = 'https://drive.google.com/uc?export=download&id='+URL_films.split('/')[-2]

URL_rating  = 'https://drive.google.com/file/d/1y5qOtma-TmwIAwVfAz6RDmy50_2r10wf/view?usp=sharing'
rating_file = 'https://drive.google.com/uc?export=download&id='+URL_rating.split('/')[-2]
```

```python
# charger les fichiers dans deux dataframes: movies et rating
movies = pd.read_excel(movies_file)
rating = pd.read_csv(rating_file, sep=';')
```

```python
# Afficher les cinq premières lignes
movies.head(5)
```

```python
# Afficher les cinq premières lignes
rating.head(5)
```

### Concaténation

Créer un dataframe contenant les deux premiers en faisant la concaténation selon l’axe 1. Que remarquez-vous? Proposez une alternative

```python
# faire une concatenation sur l'axe 1
df_concat = pd.concat([movies, rating], axis = 1)
df_concat
```

En faisant la concaténation des deux dataframes, on remarque que les id des films ne correspondent pas, c’est du au fait que la concaténation fait juste un empilement des dataframes selon l'axe 1.
Une alternative serait de mettre les Id des films qui son unique en index, ainsi, on aura la bonne correspondance entre le film et son évaluation.

```python
movies2 = movies.set_index("FilmId", drop=False)
```

```python
rating2 = rating.set_index("MoviesID", drop=False)
```

```python
df_concat2 = pd.concat([movies2, rating2], axis = 1)
df_concat2
```

### merge

* Créer un dataframe contenant les deux premiers en faisant un merge des deux dataframes
* Comme les noms des colonnes sont différents, il nous faut préciser le nom de la colonne à gauche et à droite pour le merge

```python
df_merge = pd.merge(movies, rating, left_on = "FilmId", right_on="MoviesID")
df_merge
```

## pratique 2

On dispose de deux fichiers sur des tests physico-chimiques pour des vins rouges et blancs

```python
# Importer pandas
import pandas as pd
```

```python
# charger les données contenues dans un fichiers excel ou csv se trouvant sur drive
URL_vins_rouges  = 'https://drive.google.com/file/d/1Tthug9VrRDVH9aHQhJPmkjdKGCvsSZJt/view?usp=sharing'
vins_rouges_file = 'https://drive.google.com/uc?export=download&id='+URL_vins_rouges.split('/')[-2]

URL_vins_blancs  = 'https://drive.google.com/file/d/16t4CBHsO_MNMxVqZZdJfD3WzfCa4QdoE/view?usp=sharing'
vins_blancs_file = 'https://drive.google.com/uc?export=download&id='+URL_vins_blancs.split('/')[-2]
```

```python
# charger les fichiers dans deux dataframes: movies et rating
vins_rouges = pd.read_csv(vins_rouges_file, sep=";")
vins_blancs = pd.read_csv(vins_blancs_file, sep=";")
```

```python
# Dans chaque dataframe, créer une colonne pour indiquer le type de vin
vins_rouges["type"] = "rouge"
vins_blancs["type"] = "blanc"
```

```python
# selectionner et ne garder que les colonnes: fixed acidity, citric acid, density et type dans chaque dataframes
vins_rouges = vins_rouges.loc[:, ["fixed acidity", "citric acid", "density", "type", "quality"]]
vins_blancs = vins_blancs.loc[:, ["fixed acidity", "citric acid", "density", "type", "quality"]]
```

```python
# Faire une concatenation des deux datafrmaes
df = pd.concat([vins_rouges, vins_blancs], ignore_index = True)
df
```

```python
# Effectuer un shuffle des données pour avoir un dataframe aleatoire
df = df.sample(frac= 1, random_state=1).reset_index(drop=True)
df
```
